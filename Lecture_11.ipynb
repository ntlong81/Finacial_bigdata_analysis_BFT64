```python
# Dữ liệu
du_lieu = [25, 35, 20, 10, 10]
nhan = ['A', 'B', 'C', 'D', 'E']

# Tạo figure và axes
fig, ax = plt.subplots()

# Nổi bật phần nhãn "B"
noi_bat = [0, 0.1, 0, 0, 0]  # Tạo danh sách nổi bật, giá trị 0.1 làm phần "B" nổi bật hơn

# Vẽ biểu đồ hình tròn
ax.pie(du_lieu, labels=nhan, autopct='%1.1f%%', startangle=90, explode=noi_bat)

# Thiết lập tiêu đề
ax.set_title("Ví dụ biểu đồ hình tròn đơn giản (Nổi bật B)")

# Hiển thị biểu đồ
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

# Dữ liệu ví dụ
labels = ['Giao tiếp', 'Kỹ thuật', 'Sáng tạo', 'Quản lý thời gian', 'Làm việc nhóm']
num_vars = len(labels)

# Dữ liệu của một cá nhân
values = [7, 8, 6, 9, 7]

# Đóng vòng tròn
values += values[:1]
angles = np.linspace(0, 2 * np.pi, num_vars, endpoint=False).tolist()
angles += angles[:1]

# Tạo figure và axis
fig, ax = plt.subplots(figsize=(6, 6), subplot_kw=dict(polar=True))

# Vẽ biểu đồ radar
ax.plot(angles, values, linewidth=2, linestyle='solid')
ax.fill(angles, values, alpha=0.25)

# Thiết lập nhãn
ax.set_xticks(angles[:-1])
ax.set_xticklabels(labels)

# Thiết lập phạm vi giá trị
ax.set_ylim(0, 10)

# Tiêu đề
ax.set_title('Kỹ năng cá nhân', size=16, pad=20);
```

```python
df = pd.read_csv('Data\Data_DS_Handbook\california_cities.csv')
df.head()
```

```python
# Xử lý dữ liệu: loại bỏ các giá trị thiếu trong các cột cần thiết
df_clean = df[['latd', 'longd', 'population_total', 'area_total_km2']].dropna()

# Tính log dân số để dùng cho màu
df_clean['log_pop'] = np.log10(df_clean['population_total'])

# Tạo figure và axes theo OO interface
fig, ax = plt.subplots(figsize=(10, 6))

# Vẽ biểu đồ scatter
sc = ax.scatter(
    df_clean['longd'],              # trục x
    df_clean['latd'],               # trục y
    s=df_clean['area_total_km2'],  # scale diện tích cho dễ thấy
    c=df_clean['log_pop'],          # màu theo log(pop)
    cmap='viridis',                 # colormap
    alpha=0.7,                      # độ trong suốt
    edgecolors='k',                 # viền đen cho marker
    marker='o'                      # hình tròn
)

# Thêm colorbar
cbar = fig.colorbar(sc, ax=ax)
cbar.set_label('log(Population)', fontsize=12)

# Gán nhãn trục
ax.set_xlabel('Longitude')
ax.set_ylabel('Latitude')
ax.set_title('Scatter Chart of Cities by Area and Population (log scale)', fontsize=14)

plt.tight_layout()
plt.show()
```
