```python
def normalize(text):
    text_norm = text.lower()
    special_letters = {'.',',','-','*','!',"?",':'}
    for letter in special_letters:
        text_norm = text_norm.replace(letter,' ')
    while '  ' in text_norm:
        text_norm = text_norm.replace("  "," ")
    return text_norm
```

```python
normalize('aa:. sdf! sdfsf?')
```

```python
def get_fre_word(text):
    fre_dict = {}
    for word in text.split():
        if word in fre_dict:
            fre_dict[word] += 1
        else:
            fre_dict[word] = 1
    return fre_dict
```
