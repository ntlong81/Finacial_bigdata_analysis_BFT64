### Lecture 12: Visualization - Seaborn

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Data\Pokemon.csv")
df.head()
```

```python
df.columns
```

```python
sns.pairplot(data = df[['Type 2', 'Total', 'HP', 'Attack', 'Defense']],
             hue='Type 2');
```

```python
ax = sns.histplot(data = df['HP'],
             kde=True)
ax.set_title('Histogram of Hp distribution')
ax.set_xlabel('New x label')
plt.xticks(rotation=90);
```

```python
sns.jointplot(data=df,
              x='Attack', 
              y='Defense');
```

```python
df.head()
```

```python
df_clean = df.dropna(subset=['Type 2'])

# V biu 
sns.jointplot(
    data=df_clean,
    x='Attack',
    y='Defense',
    hue='Type 2',
    kind='scatter',
    palette='Set2'
);
```

```python
sns.get_dataset_names()
```

```python
dm = sns.load_dataset('diamonds')
dm.head()
```

```python
dm.drop(columns=['cut','color', 'clarity'],
        inplace=True)
dm.head()
```

```python
# Gi s bn  c DataFrame tn l dm
corr = dm.corr(numeric_only=True)

# To mask cho tam gic di
mask = np.tril(np.ones_like(corr, dtype=bool))

# V heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, mask=mask, annot=True, cmap='coolwarm', square=True, linewidths=0.5, fmt=".2f")
plt.title('Heatmap ma trn tng quan (ch na trn)')
plt.show()
```

```python
# Gi s DataFrame ca bn tn l dm
corr = dm.corr()

# To mask ch cho phn tam gic di (khng bao gm ng cho chnh)
mask = np.tril(np.ones_like(corr, dtype=bool), k=-1)

# V heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(corr, mask=mask, annot=True, cmap='Blues', square=True, linewidths=0.5, fmt=".2f")
plt.title('Heatmap ma trn tng quan ')
plt.show()
```
